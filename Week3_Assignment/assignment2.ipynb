#### 5. Sampling (8 points)
Use simple random sampling without replacement to generate a subsampled dataset with an equal number of majority class and minority class instances, i.e., equal number of instances without and with hypothyroidism. How many majority and minority class instances are there now? Use box plots to assess how representative the subsampled majority class instances are of the majority class.

```python
############################
## GENERATED USING ChatGPT##
############################

# Identify majority and minority classes
majority_class = data['Class'].value_counts().idxmax()
minority_class = data['Class'].value_counts().idxmin()

# Number of instances in the minority class
n_minority = data['Class'].value_counts().min()

# Perform random sampling without replacement for majority class
majority_class_data = data[data['Class'] == majority_class]
minority_class_data = data[data['Class'] == minority_class]

subsampled_majority_class_data = majority_class_data.sample(n=n_minority, random_state=42)

# Combine the subsampled majority class with the minority class to create a balanced dataset
balanced_data = pd.concat([subsampled_majority_class_data, minority_class_data])

# Output the number of majority and minority class instances in the subsampled dataset
print(f'Number of majority class instances: {len(subsampled_majority_class_data)}')
print(f'Number of minority class instances: {len(minority_class_data)}')

# Create individual box plots to compare original and subsampled majority class instances
for feature in continuous_features:
    plt.figure(figsize=(10, 5))
    plt.boxplot([majority_class_data[feature].dropna(), subsampled_majority_class_data[feature].dropna()],
                labels=['Original Majority', 'Subsampled Majority'])
    plt.title(f'Box Plot of Original vs Subsampled Majority Class for {feature}')
    plt.xlabel('Class')
    plt.ylabel(f'Standardized {feature}')
    plt.grid(True)
    plt.show()
```

Overall the subsampled distributions for Age, TSH, T3, TT4, T4U, FTI have a very similar distribution to the original majority data, while reducing the overal quantity of outliers. The TBG subsampled majority does have a shifted median upwards with a reduced quantity of outliers.


#### Extra Credit. PCA (3 points)
Use principal component analysis (PCA) to reduce the dataset to 2 features (dimensions). Visualize the results. What do you observe?

```python
############################
## GENERATED USING ChatGPT##
############################

from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer

# Handling missing values (NaNs), detemrine continous features
imputer = SimpleImputer(strategy='mean')
data[continuous_features] = imputer.fit_transform(data[continuous_features])

# Insert code here.# Apply PCA to reduce the dataset to 2 dimensions
pca = PCA(n_components=2)
pca_result = pca.fit_transform(data[continuous_features])

# Create a DataFrame with PCA results
pca_df = pd.DataFrame(data=pca_result, columns=['PCA1', 'PCA2'])
pca_df['Class'] = data['Class'].values

#Determine combination of principal components.
principal_components = pca.components_
pc_df = pd.DataFrame(principal_components, columns=data[continuous_features].columns, index=['PCA1', 'PCA2'])
print("Principal Components (Linear Combinations of Features):")
print(pc_df)

# Visualize the PCA results
plt.figure(figsize=(10, 7))
for class_value in pca_df['Class'].unique():
    subset = pca_df[pca_df['Class'] == class_value]
    plt.scatter(subset['PCA1'], subset['PCA2'], label=class_value, alpha=0.5)

plt.title('PCA of Hypothyroid Disease Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Class')
plt.grid(True)
plt.show()
```

Overall, principal component analysis is highly sucessfull at determining the linear combination of the continous features that allows for the effective separation of the hypothyroid and negative clases. Overall we note that TT4, FTI, and T3 dominate the influence of principal component 1, whilst T4U, FTI, and Age dominate the influence of principal component 2. As shown in the plot above, it would be possible to  easily determine a well decision boundary for these features based on these two principal components.
